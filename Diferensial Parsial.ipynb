# Persamaan Diferensial Parsial

#### Menyelesaikan permasalahan Persamaan Diferensial Biasa Damped Harmonic Oscillation menggunakan Python Scipy

```python
# import library
from scipy.integrate import odeint, ode
import numpy as np
import matplotlib.pyplot as plt

np.set_printoptions(suppress=True, precision=10)
```

#### Selanjutnya kita buat fungsi right hand side untuk mencari dp/dt

```python
def dy(y, t, zeta, w0):
    
    x, p = y[0], y[1]
    
    dx = p
    dp = -2 * zeta * w0 * p - w0**2 * x

    return [dx, dp]
```

#### selanjutnya kita atur agar posisi awal oscillator di y = 1.0 dan kecepatanya dp/dt = 0.0, stelah itu kita buat variabel waktu sebanyak 1000 titik dengan nilai antara 0 sampai 10 menggunakan fungsi np.linespace , dan tidak lupa kita definisikan w0 system.

```python
# initial state: 
y0 = [1.0, 0.0]

# time coordinate to solve the ODE for
t = np.linspace(0, 10, 1000)
w0 = 2*np.pi*1.0
```

#### selanjutnya kita selesaikan ODE dengan fungsi odeint dari module scipy

```python
# solve the ODE problem for three different values of the damping ratio
y1 = odeint(dy, y0, t, args=(0.0, w0)) # undamped
y2 = odeint(dy, y0, t, args=(0.1, w0)) # under damped
y3 = odeint(dy, y0, t, args=(1.1, w0)) # critial damping
y4 = odeint(dy, y0, t, args=(5.0, w0)) # over damped
```

#### Nah sekarang kita plot hasilnya kedalam grafik menggunakan library matplotlib

```python
fig, ax = plt.subplots()
ax.plot(t[:500], y1[:500,0], 'k', label="undamped", linewidth=0.25)
ax.plot(t[:500], y2[:500,0], 'r', label="under damped")
ax.plot(t[:500], y3[:500,0], 'b', label=r"critical damping")
ax.plot(t[:500], y4[:500,0], 'g', label="over damped")
ax.legend();
```
